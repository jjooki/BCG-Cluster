```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import astropy.io.fits as fits
```

```python
data_path = './data/dc_data/'
for file in sorted(os.listdir(data_path)):
    print(file)
```

```python
image_data = fits.open(data_path + os.listdir(data_path)[0])
image_data.info()
```

```python
image = image_data[0].data
header = image_data[0].header
print(image.shape)
if image.ndim > 2:
    image = image.reshape(image.shape[0],-1)
print(image.shape)
max_value = np.percentile(image,99.8)
min_value = np.percentile(image,15)

plt.figure(figsize=(8,6))
plt.imshow(image, cmap='gray', vmax = max_value, vmin = min_value, origin = 'lower')
# plt.xlim(1000,1750)
# plt.ylim(300,1000)
plt.show()
```
